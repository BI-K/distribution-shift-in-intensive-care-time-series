## Clean Samples from unplausible values (>500)

```python
import torch
import numpy as np
```

```python
def clean_data_from_samples_with_gigantic_values(input_path):
    data = torch.load(input_path, weights_only=False)
    samples = data["samples"]
    samples = torch.tensor(samples, dtype=torch.float64)
    labels = data["labels"]
    labels = torch.tensor(labels, dtype=torch.float64)
    cleaned_samples = []
    cleaned_labels = []
    excluded_samples = 0
    for entry, label in zip(samples, labels):
        # think of mroe sophisticated cleaning - here we still get weird values for spo2 and nbp
        # check if any element in entry or label is greater than 500
        # flatten entry to 1D
        # cvp
        if torch.any( -15 > entry[:,0]) or torch.any(entry[:,0] > 50) or torch.any(label[:,0] > 50) or torch.any(-15 > label[:,0]):
            excluded_samples += 1
            continue
        # hr
        if torch.any( 10 > entry[:,1]) or torch.any(entry[:,1] > 330) or torch.any(label[:,1] > 330) or torch.any(10 > label[:,1]):
            excluded_samples += 1
            continue
        # nbp sys
        if torch.any( 8 > entry[:,2]) or torch.any(entry[:,2] > 350) or torch.any(label[:,2] > 350) or torch.any(8 > label[:,2]):
            excluded_samples += 1
            continue
        # nbp dias
        if torch.any( 8 > entry[:,3]) or torch.any(entry[:,3] > 150) or torch.any(label[:,3] > 150) or torch.any(8 > label[:,3]):
            excluded_samples += 1
            continue
        # nbp mean
        if torch.any( 8 > entry[:,4]) or torch.any(entry[:,4] > 170) or torch.any(label[:,4] > 170) or torch.any(8 > label[:,4]):
            excluded_samples += 1
            continue
        # spo2
        if torch.any( 40 > entry[:,5]) or torch.any(entry[:,5] > 100) or torch.any(label[:,5] > 100) or torch.any(40 > label[:,5]):
            excluded_samples += 1
            continue
        else:
            cleaned_samples.append(entry)
            cleaned_labels.append(label)

    cleaned_samples = np.array(cleaned_samples)
    cleaned_samples = torch.from_numpy(cleaned_samples)

    data["samples"] = cleaned_samples
    data["labels"] = np.array(cleaned_labels)
    output_path = input_path.replace(".pt", "_cleaned.pt")
    torch.save(data, output_path)

    print(f"{input_path}: Excluded {excluded_samples} samples with implausible values")
```

```python
data_paths = [#"test_cardiac_surgery.pt", "test_no_cardiac_surgery.pt", "train_cardiac_surgery.pt", "train_no_cardiac_surgery.pt",
              "test_ventilation.pt", "test_no_ventilation.pt", "train_ventilation.pt", "train_no_ventilation.pt",
              "test_vasopressors.pt", "test_no_vasopressors.pt", "train_vasopressors.pt", "train_no_vasopressors.pt",
             # "test_respiratory_surgery.pt", "test_no_respiratory_surgery.pt", "train_respiratory_surgery.pt", "train_no_respiratory_surgery.pt"
             ]

for path in data_paths:
    clean_data_from_samples_with_gigantic_values(path)
```

## Transform samples to predict only one channel (SpO2 and NBP Mean) and into classification tasks (hypoxemia + hypotension)

```python
#ventilation_train_label_spo2 = input_data["labels"][:,0,5]
#ventilation_train_label_nbpmean = input_data["labels"][:,0,4]

def reduce_labels_to_channel_at_position(input_path, position, label):
    input_data = torch.load(input_path, weights_only=False)
    print(input_data["samples"].shape)
    print(input_data["labels"][:5])
    input_data["labels"] = input_data["labels"][:, 0, position]
    # wrap each element in input_data["labels"] in a list
    input_data["labels"] = [[[[x.item()]]] for x in input_data["labels"]]
    input_data["labels"] = np.array(input_data["labels"])
    output_path = input_path.replace(".pt", f"_{label}.pt")
    torch.save(input_data, output_path)
```

```python
def label_as_hypoxemia(input_path):
    if input_path.endswith("_spo2.pt"):
        input_data = torch.load(input_path, weights_only=False)
        input_data["labels"] = [1 if x[0][0] <= 92 else 0 for x in input_data["labels"]]
        input_data["labels"] = np.array(input_data["labels"])
        output_path = input_path.replace("_spo2.pt", "_hypoxemia.pt")
        torch.save(input_data, output_path)
    else:
        print("A non-spo2 file was processed.")

def label_as_hypotension(input_path):
    if input_path.endswith("_nbp_mean.pt"):
        input_data = torch.load(input_path, weights_only=False)
        input_data["labels"] = [1 if x[0][0] <= 65 else 0 for x in input_data["labels"]]
        input_data["labels"] = np.array(input_data["labels"])
        output_path = input_path.replace("_nbp_mean.pt", "_hypotension.pt")
        torch.save(input_data, output_path)
    else:
        print("A non-nbp_mean file was processed.")
```

```python
data_paths = [# "test_cardiac_surgery_cleaned.pt", "test_no_cardiac_surgery_cleaned.pt", "train_cardiac_surgery_cleaned.pt", "train_no_cardiac_surgery_cleaned.pt",
               "test_ventilation_cleaned.pt", "test_no_ventilation_cleaned.pt", "train_ventilation_cleaned.pt", "train_no_ventilation_cleaned.pt",
               "test_vasopressors_cleaned.pt", "test_no_vasopressors_cleaned.pt", "train_vasopressors_cleaned.pt", "train_no_vasopressors_cleaned.pt",
              # "test_respiratory_surgery_cleaned.pt", "test_no_respiratory_surgery_cleaned.pt", "train_respiratory_surgery_cleaned.pt", "train_no_respiratory_surgery_cleaned.pt"
              ]
```

```python
# for spo2
for data_path in data_paths:
    reduce_labels_to_channel_at_position(data_path, 5, "spo2")
```

```python
# for nbp mean
for data_path in data_paths:
    reduce_labels_to_channel_at_position(data_path, 4, "nbp_mean")
```

```python
for data_path in data_paths:
    data_path = data_path.replace(".pt", "_spo2.pt")
    label_as_hypoxemia(data_path)
    data_path = data_path.replace("_spo2.pt", "_nbp_mean.pt")
    label_as_hypotension(data_path)
```
